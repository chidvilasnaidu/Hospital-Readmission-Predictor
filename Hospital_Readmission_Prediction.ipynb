```python
import optuna
from sklearn.model_selection import StratifiedKFold
from optuna.samplers import TPESampler
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
```

# Hyper Parameter Tuning 

### Using `accuracy score` Metrics 

```python
def Objective(trail):
    model=trail.suggest_categorical('model',['KNN','Decision_Tree','Logistic_Regression','Random_Forest','XGBoost'])
    if model=='KNN':
        n_neighbors=trail.suggest_int('n_neighbors',1,10)
        weights=trail.suggest_categorical('weights',['uniform','distance'])
        OBJ=KNeighborsClassifier(n_neighbors=n_neighbors,weights=weights)
    elif model=='Decision_Tree':
        criterion=trail.suggest_categorical('criterion',["gini","entropy","log_loss"])
        dt_min_samples_leaf=trail.suggest_int('dt_min_samples_leaf',2,10)
        max_leaf_nodes=trail.suggest_int('max_leaf_nodes',2,10)
        dt_min_samples_split=trail.suggest_int('dt_min_samples_split',2,10)
        dt_max_depth=trail.suggest_int('dt_max_depth',2,10)
        OBJ=DecisionTreeClassifier(criterion=criterion,min_samples_leaf=dt_min_samples_leaf,max_leaf_nodes=max_leaf_nodes,min_samples_split=dt_min_samples_split,max_depth=dt_max_depth)
    elif model=='Logistic_Regression':
        l1_ratio=trail.suggest_float('l1_ratio',0.0,1.0)
        lr_C=trail.suggest_float('lr_C',1e-4,100,log=True)
        solver=trail.suggest_categorical('solver',['lbfgs','newton-cg','saga'])
        lr_max_iter=trail.suggest_int('lr_max_iter',200,2000)
        lr_class_weight=trail.suggest_categorical('lr_class_weight',[None,'balanced'])
        OBJ = LogisticRegression(C=lr_C, solver=solver, max_iter=lr_max_iter, class_weight=lr_class_weight, random_state=42 )
    elif model=='Random_Forest':
        rf_n_estimators=trail.suggest_int('rf_n_estimators',100,1000)
        rf_max_depth=trail.suggest_int('rf_max_depth',2,50)
        rf_min_samples_split=trail.suggest_int('rf_min_samples_split',2,50)
        rf_min_samples_leaf=trail.suggest_int('rf_min_samples_leaf',1,20)
        max_features=trail.suggest_categorical('max_features',['sqrt','log2',None])
        bootstrap=trail.suggest_categorical('bootstrap',[True,False])
        rf_class_weight=trail.suggest_categorical('rf_class_weight',[None,'balanced','balanced_subsample'])
        OBJ=RandomForestClassifier(n_estimators=rf_n_estimators,max_depth=rf_max_depth,min_samples_split=rf_min_samples_split,min_samples_leaf=rf_min_samples_leaf,max_features=max_features,bootstrap=bootstrap,class_weight=rf_class_weight,random_state=42,n_jobs=-1)
    elif model=='XGBoost':
        xgb_n_estimators=trail.suggest_int('xgb_n_estimators',100,1000)
        xgb_max_depth=trail.suggest_int('xgb_max_depth',2,15)
        learning_rate=trail.suggest_float('learning_rate',0.001,0.3,log=True)
        subsample=trail.suggest_float('subsample',0.5,1.0)
        colsample_bytree=trail.suggest_float('colsample_bytree',0.5,1.0)
        xgb_gamma=trail.suggest_float('xgb_gamma',0.0,5.0)
        xgb_min_child_weight=trail.suggest_int('xgb_min_child_weight',1,10)
        reg_alpha=trail.suggest_float('reg_alpha',1e-8,10,log=True)
        reg_lambda=trail.suggest_float('reg_lambda',1e-8,10,log=True)
        OBJ=XGBClassifier(n_estimators=xgb_n_estimators,max_depth=xgb_max_depth,learning_rate=learning_rate,subsample=subsample,colsample_bytree=colsample_bytree,gamma=xgb_gamma,min_child_weight=xgb_min_child_weight,reg_alpha=reg_alpha,reg_lambda=reg_lambda,random_state=42,n_jobs=-1,eval_metric='logloss')

    cv1=StratifiedKFold(n_splits=8,shuffle=True,random_state=42)
    score=cross_val_score(estimator=OBJ,X=X_train_transformed,y=y_train_transformed,scoring='accuracy',cv=cv1).mean()
    return score
```

```python
## Create a study 
study=optuna.create_study(direction='maximize',  study_name='Baysenj11',sampler=TPESampler())
```

```python
## Optimize

study.optimize(Objective,n_trials=100)
```

```python

print('Best Score :',study.best_value)
print('Best Params:',study.best_params)
```
